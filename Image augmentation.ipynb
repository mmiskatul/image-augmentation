```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize 
```

```python
img = mpimg.imread('cat3.jpg')
```

```python
plt.imshow(img)
```

```python
img_rescale =resize(img,(800,800))
plt.imshow(img_rescale)
```

```python
img_rescale2 =resize(img,(150,150))
plt.imshow(img_rescale2)
```

```python
img_rescale3 =resize(img,(50,50))
plt.imshow(img_rescale3)
```

```python
horizontal_flip =np.fliplr(img ) 
```

```python
plt.imshow(horizontal_flip) 
```

```python
vartical_flip =np.flipud(img) 
plt.imshow(vartical_flip)
```

```python
from skimage import transform 
tran_img =transform.rotate(img,random.uniform(-40,40)) 
plt.imshow(tran_img)
```

```python
img_noise  =random_noise(img,mode='s&p',clip=True) 
plt.imshow(img_noise)
```
